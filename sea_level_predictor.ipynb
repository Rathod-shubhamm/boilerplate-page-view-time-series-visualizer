```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Untitled folder/epa-sea-level.csv')
plt.scatter(df["Year"] , df["CSIRO Adjusted Sea Level"])
```

```python
lineA = linregress(df["Year"] , df["CSIRO Adjusted Sea Level"])
xA = np.arange(df["Year"].min() ,2050 ,1)
yA = xA * lineA.slope + lineA.intercept
plt.plot(xA , yA)
```

```python
df_2000 = df[df["Year"] >= 2000]

# Perform linear regression on the filtered data
lineB = linregress(df_2000["Year"], df_2000["CSIRO Adjusted Sea Level"])

# Generate x and y values for the regression line
xB = np.arange(2000, 2050, 1)
yB = xB * lineB.slope + lineB.intercept

# Plot the actual data points
plt.scatter(df_2000["Year"], df_2000["CSIRO Adjusted Sea Level"], label="Data", color='blue')

# Plot the regression line
plt.plot(xB, yB, label="Best Fit Line (2000-2050)", color='red')

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Sea Level (inches)")
plt.title("Rise in Sea Level (2000-2050)")

# Add legend
plt.legend()

# Save the plot as a PNG file
plt.savefig("Sea_level_plot.png")

# Display the plot
plt.show()
```
